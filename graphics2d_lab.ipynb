### Lab9:  Rotation And Scale Transformation

In this lab, your objective is to implement the rotation and scale  transformation for graphics. You should complete the `Transform2DRotation` and  `Transform2DScale` class within `Graphics2D/algorithm.py` file.

#### references
 - [åéä¸çº¿æ§ä»£æ°](https://sites.cs.ucsb.edu/~lingqi/teaching/resources/GAMES101_Lecture_02.pdf)
 - [åæ¢ï¼äºç»´ä¸ä¸ç»´ï¼](https://sites.cs.ucsb.edu/~lingqi/teaching/resources/GAMES101_Lecture_03.pdf)



````python
from Graphics2D.transform import Transform2DRotation

plt.figure(figsize=(12.8,4.8))
ax1=plt.subplot(1,2,1)
ax2=plt.subplot(1,2,2)

canvas.unmount_shapes()
canvas.add_shape(
        Circle2D((500, 300), 200)
)
image=canvas.show(filling_color=(0,255,0))
ax1.imshow(image)

canvas.unmount_shapes()
transform2d=Transform2DRotation(theta=np.pi/10)
circle=Circle2D((500, 300), 200)
circle.transform(transform2d)
canvas.add_shape(
     circle
)
image=canvas.show(filling_color=(0,255,0))
ax2.imshow(image)
````

````python
from Graphics2D.transform import Transform2DScale

plt.figure(figsize=(12.8,4.8))
ax1=plt.subplot(1,2,1)
ax2=plt.subplot(1,2,2)

canvas.unmount_shapes()
canvas.add_shape(
        Circle2D((500, 300), 200)
)
image=canvas.show(filling_color=(0,255,0))
ax1.imshow(image)

canvas.unmount_shapes()
transform2d=Transform2DScale(1.5)
circle=Circle2D((500, 300), 200)
circle.transform(transform2d)
canvas.add_shape(
     circle
)
image=canvas.show(filling_color=(0,255,0))
ax2.imshow(image)
````

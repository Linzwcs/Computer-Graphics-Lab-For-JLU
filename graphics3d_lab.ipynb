### Lab10-12: 

In this lab, you should complete function `model_transform_matrix` `perspective_projection_matrix` within the `Graphics3D/algorithm.py` file. After that, you need to undestand  `VertexProcess` class `Graphics3D/VertexProcess.py` and  `Rasterizer` class in `Graphics3D/Rasterizer.py`

#### references:

- [变换（模型、视图、投影）](https://sites.cs.ucsb.edu/~lingqi/teaching/resources/GAMES101_Lecture_04.pdf).

```python
from Graphics3D import (
    Shader,
    Object3D,
    VertexProcess,
    FaceDefaultProcess,
    MVP,
    TextureDefaultProcess,
    TextureRandomColorProcess,
    TextureImageProcess,
    BaseLighting,
    BlinnPhongLighting,
    Rasterizer,
    DefaultRender, 
    BlinnPhongRender,
    load_obj
)
import numpy as np
from matplotlib import  pyplot as plt
```

```python
shader=Shader()
shader.set_MVP_processor(MVP([0, 0, 60, 1], [0, 1, 0, 0], [0, 0, -1, 0]))
shader.set_vertexs_processor(VertexProcess())
shader.set_face_processor(FaceDefaultProcess())
shader.set_rasterizer_processor(Rasterizer())
shader.set_texture_mapping_processor(TextureRandomColorProcess())
shader.set_lighting_processor(BaseLighting())
shader.set_render_processor(DefaultRender())
args=load_obj('./asserts/Basketball/basketball.obj')
image=shader.display(Object3D(**args))
plt.figure(figsize=(10,6))
plt.imshow(image)
```
